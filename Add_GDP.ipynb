```python
import numpy as np
import os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
file_name = 'Country_Area'
df = pd.read_csv(f'Train/{file_name}.csv')
df.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
df
```

```python
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)_per_capita#UN_estimates_between_1970_and_1979')
```

```python
gdp_name = table[1]['Country (or dependent territory)'].unique()
```

```python
gdp = pd.DataFrame(columns = ['Country (or dependent territory)','GDP','Year'])
gdp
```

```python
for t in table[1:5]:
    years = t.columns.drop('Country (or dependent territory)')
    for i in years:
        t['Year'] = float(i)
        t['GDP'] = t[i]
        x = t[['Country (or dependent territory)','GDP','Year']]
        gdp = gdp.append(x)
```

```python
t = table[5]
years = ['2020','2021']
for i in years:
    t['Year'] = float(i)
    t['GDP'] = t[i]
    x = t[['Country (or dependent territory)','GDP','Year']]
    gdp = gdp.append(x)
```

```python
i = '2022'
table[5]['Year'] = float(i)
table[5]['GDP'] = table[5][i]
save_to_test = table[5][['Country (or dependent territory)','GDP','Year']]
save_to_test.rename(columns = {'Country (or dependent territory)':
                               'Region, subregion, country or area *'}, inplace = True)
save_to_test.replace('—',np.nan,inplace = True)
```

```python
gdp.rename(columns = {'Country (or dependent territory)':'Region, subregion, country or area *'}, inplace = True)
```

```python
gdp.replace('—',np.nan,inplace = True)
```

```python
x = pd.merge(gdp, df[df['Year'] >= 1980], on =['Region, subregion, country or area *', 'Year'])
```

```python
x
```

```python
save_path = './Data/Train'
file_name = 'set_with_gdp'
x.to_csv(f'{save_path}/{file_name}.csv') 
```
